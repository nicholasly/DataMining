```python
import pandas as pd
import numpy as np
from sklearn import tree
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn import preprocessing
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train.csv', sep=',', header=None)
print(data.shape)
data.head()
```

```python
feature_list = []
for i in range(1,129):
    feature_list.append(i)
X = data[feature_list]
X.tail()
```

```python
label = 129
Y = data[label]
Y.tail()
```

```python
min_max_scaler = preprocessing.MinMaxScaler()
X_scaled = min_max_scaler.fit_transform(X)
X_scaled.shape
```

```python
clf = RandomForestClassifier(n_estimators=200, oob_score=True, random_state=10)
#clf = AdaBoostClassifier(n_estimators=1000)
clf.fit(X, Y)
clf.score(X, Y)
```

```python
p_data = pd.read_csv('test_raw.csv', sep=',', header=None)
print(p_data.shape)
p_data.head()
```

```python
p_X = p_data[feature_list]
min_max_scaler = preprocessing.MinMaxScaler()
p_X_scaled = min_max_scaler.fit_transform(p_X)
p_X_scaled.shape
```

```python
p_Y = clf.predict(p_X)
```

```python
ID = p_data[0]
print(ID)
```

```python
import csv
with open("submission03.csv","w") as csvfile: 
    writer = csv.writer(csvfile, lineterminator='\n')
    writer.writerow(["ID","Pred"])
    for i in range(0,7910):
        writer.writerow([ID[i], ''.join(['cls_', str(p_Y[i])])])
```
